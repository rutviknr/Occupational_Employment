Step-3: Data Preparation

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns=['Occupation Name'],drop_first=True)
```

```python
df
```

```python
df.isnull().sum()
```

Step-4: Normalisation

```python
X = df.drop(columns=['Percent Change','State FIPS','Area Name','Occupation Code'])
y = df['Percent Change']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

Step-5: Building the Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train_trf.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train_trf, y_train, epochs=100, batch_size=32, verbose=1)
```

Step-6: Evaluating the Model

```python
mse = model.evaluate(X_test_trf, y_test)
print(f'Mean Squared Error: {mse}')
```
